# Bidirectional Recurrent Neural Network (BiRNN)

## 1. Introduction

A **Bidirectional Recurrent Neural Network (BiRNN)** is an extension of the standard RNN that processes sequential data in **both directions**:

* **Forward Direction:** Past → Future
* **Backward Direction:** Future → Past

The outputs from both directions are combined to create a richer representation of each time step.

$$
\text{BiRNN} = \text{Forward RNN} + \text{Backward RNN}
$$

---

## 2. Why Bidirectional RNN?

A standard RNN processes a sequence only from left to right.

$$
x_1 \rightarrow x_2 \rightarrow x_3 \rightarrow \cdots \rightarrow x_T
$$

Therefore, the hidden state at time step $t$ contains information only from the **current and previous inputs**.

$$
h_t = f(x_1, x_2, \ldots, x_t)
$$

However, in many sequence-processing tasks, understanding the current input requires information from both **past and future context**.

For example:

> "The bank approved my loan application."

The meaning of **bank** becomes clearer by considering the complete sentence.

A BiRNN solves this limitation by processing the sequence in both directions.

---

## 3. Architecture of Bidirectional RNN

A BiRNN consists of **two independent RNNs**.

### Forward RNN

The forward RNN processes the sequence from left to right.

$$
x_1 \rightarrow x_2 \rightarrow x_3 \rightarrow \cdots \rightarrow x_T
$$

The forward hidden state is calculated as:

$$
\overrightarrow{h_t}
=

f
\left(
W_{xh}^{(f)}x_t
+
W_{hh}^{(f)}\overrightarrow{h_{t-1}}
+
b_h^{(f)}
\right)
$$

The forward hidden state contains information from:

$$
x_1, x_2, \ldots, x_t
$$

### Backward RNN

The backward RNN processes the sequence from right to left.

$$
x_T \rightarrow x_{T-1} \rightarrow \cdots \rightarrow x_2 \rightarrow x_1
$$

The backward hidden state is calculated as:

$$
\overleftarrow{h_t}
=

f
\left(
W_{xh}^{(b)}x_t
+
W_{hh}^{(b)}\overleftarrow{h_{t+1}}
+
b_h^{(b)}
\right)
$$

The backward hidden state contains information from:

$$
x_T, x_{T-1}, \ldots, x_t
$$

---

## 4. Combining Forward and Backward Hidden States

At every time step, the forward and backward hidden states are combined.

The most commonly used method is **concatenation**.

$$
h_t
=

\left[
\overrightarrow{h_t}
;
\overleftarrow{h_t}
\right]
$$

Therefore, the final hidden representation contains information from both directions.

$$
h_t
=

\text{Past Context}
+
\text{Current Input}
+
\text{Future Context}
$$

If each RNN has $H$ hidden units, then:

$$
\overrightarrow{h_t} \in \mathbb{R}^{H}
$$

$$
\overleftarrow{h_t} \in \mathbb{R}^{H}
$$

After concatenation:

$$
h_t \in \mathbb{R}^{2H}
$$

---

## 5. Output Calculation

The combined hidden representation can be passed to an output layer.

$$
y_t
=

g
\left(
W_y h_t + b_y
\right)
$$

Where:

* $h_t$ = Combined bidirectional hidden state
* $W_y$ = Output weight matrix
* $b_y$ = Output bias
* $g$ = Output activation function

---

## 6. Standard RNN vs Bidirectional RNN

| Feature               | Standard RNN | Bidirectional RNN  |
| --------------------- | ------------ | ------------------ |
| Processing Direction  | Forward only | Forward + Backward |
| Past Context          | Yes          | Yes                |
| Future Context        | No           | Yes                |
| Number of RNNs        | One          | Two                |
| Output Representation | $H$          | Usually $2H$       |
| Computational Cost    | Lower        | Higher             |
| Context Understanding | Limited      | Richer             |

---

## 7. Bidirectional RNN Variants

The bidirectional architecture can be applied to different recurrent units.

$$
\text{BiRNN}
=

\text{Forward SimpleRNN}
+
\text{Backward SimpleRNN}
$$

$$
\text{BiLSTM}
=

\text{Forward LSTM}
+
\text{Backward LSTM}
$$

$$
\text{BiGRU}
=

\text{Forward GRU}
+
\text{Backward GRU}
$$

In practice, **BiLSTM** and **BiGRU** are commonly preferred because Simple RNNs can suffer from vanishing gradients.

---

## 8. Applications

Bidirectional RNNs are useful when the **complete input sequence is available**.

Common applications include:

* Named Entity Recognition (NER)
* Part-of-Speech Tagging
* Sentiment Analysis
* Text Classification
* Speech Recognition
* Sequence Labeling
* Encoder networks in Sequence-to-Sequence models

---

## 9. Limitations

### Computational Cost

A BiRNN uses two independent RNNs, which increases computation and the number of trainable parameters.

### Requires Complete Sequence

The backward RNN requires future inputs.

Therefore, standard BiRNNs are not suitable for strict real-time or causal prediction tasks where future inputs are unavailable.

### Long-Term Dependencies

A Bidirectional Simple RNN can still suffer from vanishing gradients.

BiLSTM or BiGRU architectures are commonly used for longer sequences.

---

## 10. Key Takeaways

* A BiRNN processes sequences in both forward and backward directions.
* It consists of two independent recurrent neural networks.
* The forward RNN captures past context.
* The backward RNN captures future context.
* Their hidden states are usually concatenated.
* With concatenation, $H$ hidden units per direction produce a $2H$-dimensional representation.
* BiRNNs are useful when the complete input sequence is available.
* BiLSTM and BiGRU are commonly used in practical deep learning applications.
