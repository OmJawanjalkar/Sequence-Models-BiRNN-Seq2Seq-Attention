# Encoder–Decoder Theory

## 1. Introduction

The **Encoder–Decoder** architecture is a neural network designed to transform one sequence into another sequence.

It is also known as the **Sequence-to-Sequence (Seq2Seq)** model.

Unlike traditional neural networks that produce a single output, an Encoder–Decoder model generates an entire output sequence.

**Examples:**

* English → French Translation
* Text Summarization
* Chatbots
* Question Answering
* Speech Recognition

---

# 2. Why Do We Need an Encoder–Decoder?

Traditional RNNs work well for tasks where the input and output have fixed lengths.

However, many real-world problems involve **variable-length sequences**.

For example:

```text
Input  : How are you?
Output : Comment allez-vous ?
```

The number of words in the input and output may be different.

An Encoder–Decoder architecture is designed to handle these variable-length sequence-to-sequence tasks.

---

# 3. What is an Encoder?

The **Encoder** reads the entire input sequence one token at a time.

Its job is to understand the meaning of the input and convert it into a fixed-size representation called the **Context Vector**.

```text
Input Sentence
      │
      ▼
   Encoder
      │
      ▼
Context Vector
```

---

# 4. What is a Decoder?

The **Decoder** receives the Context Vector from the Encoder.

It generates the output sequence one token at a time until the complete sequence is produced.

```text
Context Vector
      │
      ▼
   Decoder
      │
      ▼
Output Sequence
```

---

# 5. What is the Context Vector?

The **Context Vector** is the final hidden state of the Encoder.

It contains the important information learned from the input sequence.

```text
Input Sentence

↓

Encoder

↓

Context Vector

↓

Decoder

↓

Output Sentence
```

The Decoder uses this Context Vector to generate the output sequence.

---

# 6. Working of Encoder–Decoder

The complete workflow is:

```text
Input Sequence

↓

Encoder

↓

Context Vector

↓

Decoder

↓

Output Sequence
```

The Encoder first processes the input sequence.

The Decoder then generates one output token at a time until it predicts the end of the sequence.

---

# 7. Applications

Encoder–Decoder models are widely used in Natural Language Processing.

Some common applications are:

* Machine Translation
* Text Summarization
* Chatbots
* Question Answering
* Speech Recognition
* Image Captioning

---

# 8. Advantages

* Handles variable-length input and output sequences.
* Learns meaningful sequence representations.
* Suitable for many sequence-to-sequence tasks.
* Forms the foundation of modern NLP models.

---

# 9. Limitation

The basic Encoder–Decoder architecture uses a **single Context Vector** to represent the entire input sequence.

For long sentences, this fixed-size vector may not contain all the important information.

As a result, performance decreases for longer sequences.

This limitation led to the development of the **Attention Mechanism**, which allows the Decoder to focus on different parts of the input while generating each output token.

---

# 10. Key Takeaways

* Encoder–Decoder is a Sequence-to-Sequence architecture.
* The Encoder converts the input sequence into a Context Vector.
* The Decoder generates the output sequence from the Context Vector.
* It supports variable-length input and output sequences.
* The Context Vector becomes a bottleneck for long sequences.
* The Attention Mechanism was introduced to overcome this limitation.
