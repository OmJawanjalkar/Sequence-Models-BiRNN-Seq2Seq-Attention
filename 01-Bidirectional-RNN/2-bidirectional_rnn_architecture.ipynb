# Bidirectional RNN Architecture

## 1. Architecture Overview

A **Bidirectional RNN (BiRNN)** consists of two independent RNN layers:

1. **Forward RNN** — processes the sequence from left to right.
2. **Backward RNN** — processes the sequence from right to left.

$$
\text{Input Sequence}
\rightarrow
\begin{cases}
\text{Forward RNN} \
\text{Backward RNN}
\end{cases}
\rightarrow
\text{Merge Hidden States}
\rightarrow
\text{Output}
$$

---

## 2. Forward RNN

The forward RNN processes:

$$
x_1 \rightarrow x_2 \rightarrow x_3 \rightarrow \cdots \rightarrow x_T
$$

It generates forward hidden states:

$$
\overrightarrow{h_1},
\overrightarrow{h_2},
\overrightarrow{h_3},
\ldots,
\overrightarrow{h_T}
$$

The forward hidden state captures **past context**.

---

## 3. Backward RNN

The backward RNN processes:

$$
x_T \rightarrow x_{T-1} \rightarrow x_{T-2} \rightarrow \cdots \rightarrow x_1
$$

It generates backward hidden states:

$$
\overleftarrow{h_T},
\overleftarrow{h_{T-1}},
\ldots,
\overleftarrow{h_1}
$$

The backward hidden state captures **future context**.

---

## 4. Merge Hidden States

At each time step, both hidden states are combined.

The most common method is **concatenation**:

$$
h_t =
\left[
\overrightarrow{h_t};
\overleftarrow{h_t}
\right]
$$

If each RNN has $H$ hidden units:

$$
H + H = 2H
$$

So, the final representation contains information from both directions.

---

## 5. Complete Architecture Flow

$$
\boxed{
\text{Input Sequence}
\rightarrow
\text{Forward RNN + Backward RNN}
\rightarrow
\text{Merge Hidden States}
\rightarrow
\text{Output}
}
$$

In simple words:

$$
\boxed{
\text{Past Context}
+
\text{Future Context}
\rightarrow
\text{Better Sequence Representation}
}
$$
