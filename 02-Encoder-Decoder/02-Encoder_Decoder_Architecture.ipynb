# Encoder–Decoder Architecture

## 1. Architecture Overview

An **Encoder–Decoder** model consists of two main components:

* **Encoder** – Reads and understands the input sequence.
* **Decoder** – Generates the output sequence.

```text
Input Sequence
      │
      ▼
   Encoder
      │
      ▼
 Context Vector
      │
      ▼
   Decoder
      │
      ▼
Output Sequence
```

---

# 2. Encoder Architecture

The Encoder processes the input sequence one token at a time.

$$
x_1 \rightarrow x_2 \rightarrow x_3 \rightarrow \cdots \rightarrow x_T
$$

At each time step, it updates its hidden state:

$$
h_t = f(x_t, h_{t-1})
$$

where:

* $x_t$ = Current input token
* $h_{t-1}$ = Previous hidden state
* $h_t$ = Current hidden state

After processing the entire sequence, the Encoder produces its final hidden state:

$$
h_T
$$

This final hidden state becomes the **Context Vector**.

---

# 3. Context Vector

The Context Vector summarizes the information from the entire input sequence.

$$
\boxed{C = h_T}
$$

where:

* $C$ = Context Vector
* $h_T$ = Final hidden state of the Encoder

The Decoder uses this vector to generate the output sequence.

---

# 4. Decoder Architecture

The Decoder starts with the Context Vector.

At each time step, it predicts one output token.

$$
y_1 \rightarrow y_2 \rightarrow y_3 \rightarrow \cdots \rightarrow y_N
$$

The hidden state of the Decoder is computed as:

$$
s_t = f(y_{t-1}, s_{t-1})
$$

where:

* $y_{t-1}$ = Previous output token
* $s_{t-1}$ = Previous decoder hidden state
* $s_t$ = Current decoder hidden state

The output token is predicted as:

$$
y_t = g(s_t)
$$

where $g$ represents the output layer followed by the Softmax activation.

---

# 5. Complete Data Flow

```text
Input Sentence
      │
      ▼
 Embedding
      │
      ▼
   Encoder
      │
      ▼
Context Vector
      │
      ▼
   Decoder
      │
      ▼
Output Tokens
```

# 7. Architecture Summary

$$
\boxed{
\text{Input Sequence}
\rightarrow
\text{Encoder}
\rightarrow
\text{Context Vector}
\rightarrow
\text{Decoder}
\rightarrow
\text{Output Sequence}
}
$$

---

# 8. Key Points

* The Encoder reads the complete input sequence.
* The final hidden state becomes the Context Vector.
* The Context Vector is passed to the Decoder.
* The Decoder generates one output token at a time.
* The process continues until the **End-of-Sequence (EOS)** token is generated.
* The fixed-size Context Vector works well for short sequences but becomes a bottleneck for long sequences, which motivates the Attention Mechanism.
