# Practical Implemention of the  BiRNNs

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
from tensorflow.keras.datasets import imdb

max_features = 10000

(X_train, y_train), (X_test, y_test) = imdb.load_data(
    num_words=max_features
)

print("Training data shape:", X_train.shape)
print("Training labels shape:", y_train.shape)

print("Test data shape:", X_test.shape)
print("Test labels shape:", y_test.shape)
```

```python
word_index = imdb.get_word_index()
```

```python
reverse_word_index = {
    value: key
    for (key, value) in word_index.items()
}
```

```python
decoded_review = " ".join(
    reverse_word_index.get(i - 3, "?")
    for i in X_train[0]
)

print(decoded_review)
```

```python
print(len(X_train[0]))
```

```python
MAX_LEN = 200  # or 300, 500
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_train = pad_sequences(
    X_train,
    maxlen=MAX_LEN,
    padding="post",
    truncating="post"
)

X_test = pad_sequences(
    X_test,
    maxlen=MAX_LEN,
    padding="post",
    truncating="post"
)
```

```python
print("Training Shape :", X_train.shape)
print("Testing Shape  :", X_test.shape)
```

```python
print("First Review Shape :", X_train[0].shape)
print("First Label :", y_train[0])
```

```python
from tensorflow.keras.layers import Bidirectional
```

```python
VOCAB_SIZE = 10000
MAX_LEN = 200
EMBEDDING_DIM = 128
RNN_UNITS = 64
```

```python
model = Sequential([
    Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN),

   Bidirectional(
        SimpleRNN(
            units=RNN_UNITS
        )
    ),


   Dense(64, activation="relu"),
   Dense(1, activation="sigmoid")

])
```

```python
model.summary()
```

```python
model.compile(
  optimizer="adam",
  loss="binary_crossentropy",
  metrics=["accuracy"]
)
```

```python
history = model.fit(
  X_train,
  y_train,
  epochs=10,
  batch_size=32,
  validation_split=0.2
)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print(f"Test Loss     : {test_loss:.4f}")
print(f"Test Accuracy : {test_accuracy:.4f}")
```

```python
predictions = model.predict(X_test)

print(predictions[:5])
```

```python
predicted_labels = (predictions > 0.5).astype(int)

print(predicted_labels[:5])
```

```python
from tensorflow.keras.datasets import imdb

word_index = imdb.get_word_index()
```

```python
def encode_review(review):

    words = review.lower().split()

    encoded = []

    for word in words:
        if word in word_index:
            encoded.append(word_index[word] + 3)
        else:
            encoded.append(2)   # Unknown word

    return encoded
```

```python
def predict_review(review):

    encoded_review = encode_review(review)

    padded_review = pad_sequences(
        [encoded_review],
        maxlen=MAX_LEN,
        padding="post",
        truncating="post"
    )

    prediction = model.predict(padded_review, verbose=0)[0][0]

    print(f"Review : {review}")
    print(f"Probability : {prediction:.4f}")

    if prediction >= 0.5:
        print("Prediction : Positive 😊")
    else:
        print("Prediction : Negative 😞")
```

```python
predict_review("This movie was absolutely amazing and I loved every moment.")
```

```python
predict_review("The movie was boring and a complete waste of time.")
```
